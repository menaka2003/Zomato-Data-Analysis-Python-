```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
df=pd.read_csv(r"C:\Users\menak\Downloads\python\zomato_sample.csv")
```

```python
df
```

```python
# Display first few rows
df.head()
```

```python
#Display last few rows
df.tail()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Handling missing values
df['rate'] = df['rate'].fillna(df['rate'].median())
df['dish_liked'] = df['dish_liked'].fillna('Not Liked')
df['votes'] = df['votes'].fillna(0)
df
```

```python
df.info()  # Summary of the dataset
df.describe()  # Statistical summary
df.dtypes  # Check data types
```

```python
# Convert DataFrame to NumPy array
numpy_data = df.to_numpy()
print(numpy_data)
```

```python
# Check for duplicate records
duplicates = df[df.duplicated()]
print("Number of duplicate records:", len(duplicates))
```

```python
# Drop duplicates
df = df.drop_duplicates()
df
```

```python
# Value counts of city-wise restaurants
city_counts = df['listed_in(city)'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(city_counts.index[:10], city_counts.values[:10], color='skyblue')
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=40)
plt.show()
```

```python
# Rating category distribution
rating_categories = pd.cut(df['rate'], bins=[0, 2, 3, 4, 5], labels=["Poor", "Average", "Good", "Excellent"])
rating_counts = rating_categories.value_counts()
plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'pink', 'grey', 'green'])
plt.title("Distribution of Restaurants by Rating Categories")
plt.show()
```

```python
# Countplot of restaurant types
sns.countplot(y='listed_in(city)', data=df, order=df['listed_in(city)'].value_counts().index[:10])
plt.title("Top 10 Cities by Number of Restaurants")
plt.show()
```

```python
#Horizontal Bar Chart
from collections import Counter
from itertools import chain

# Split dish names and count them separately
dish_list = list(chain.from_iterable(df['dish_liked'].dropna().str.split(', ')))
dish_counts = pd.Series(Counter(dish_list)).nlargest(10)

plt.figure(figsize=(10, 6))
plt.barh(dish_counts.index, dish_counts.values, color='orange')
plt.xlabel("Count")
plt.ylabel("Dishes")
plt.title("Top 10 Most Liked Dishes")
plt.gca().invert_yaxis()  # Invert for better readability
plt.show()
```

```python
# Pie chart for online order acceptance
online_counts = df['online_order'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(online_counts, labels=online_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
plt.title("Percentage of Restaurants Accepting Online Orders")
plt.show()
```

```python
# Grouped analysis on restaurant types
grp = df.groupby('rest_type')['rate'].mean().sort_values(ascending=False)
print("Restaurant types with highest average ratings:")
print(grp.head(10))
```

```python
# Bar chart for top-rated restaurant types
plt.figure(figsize=(10,6))
plt.bar(grp.index[:10], grp.values[:10], color='purple')
plt.xticks(rotation=45)
plt.ylabel("Average Rating")
plt.title("Top 10 Restaurant Types by Average Rating")
plt.show()
```
